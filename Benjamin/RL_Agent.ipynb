# 1. Data Import

```python
load_dotenv()
API_KEY = os.getenv("API_KEY")
SECRET_KEY = os.getenv("SECRET_KEY")
```

```python
client = Client(API_KEY, SECRET_KEY)
klines = client.get_historical_klines("BTCUSDT", Client.KLINE_INTERVAL_1HOUR, "1 Jan, 2021")
data = [elt[:6] for elt in klines]
columns = ["timestamp", "Open", "High", "Low", "Close", "Volume"]
df_start = pd.DataFrame(data, columns=columns)
```

```python
btc_df = pd.DataFrame(data, columns=columns)
btc_df[columns[1:]] = btc_df[columns[1:]].astype(float)
btc_df.head()
```

```python
start = btc_df.iloc[0].timestamp.astype('datetime64[ms]')
end = btc_df.iloc[-1].timestamp.astype('datetime64[ms]')

print(f'Data from {start} until {end}')
btc_df['datetime'] = pd.to_datetime(btc_df['timestamp'],unit='ms')
btc_df
```
